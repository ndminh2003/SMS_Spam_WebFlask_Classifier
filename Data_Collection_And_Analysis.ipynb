```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
table_1 = pd.read_csv("spam.csv", encoding='windows-1252')
table_1.head()

df_table_1 = table_1.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
df_table_1.head()
```

```python
df_final_1 = df_table_1.rename(columns={"v1": "Label", "v2": "Text"}, errors="raise")
df_final_1.head()
```

```python
table_2 = pd.read_csv("spamraw.csv", encoding='utf-8')
table_2.head()
```

```python
df_final_2 = table_2.rename(columns={"type": "Label", "text": "Text"}, errors="raise")
df_final_2.head()
```

```python
df_sample = pd.concat([df_final_1, df_final_2])
df_sample.head()
```

```python
df_sample1 = df_sample.sample(frac = 1)
df_sample1.head()
```

```python
df = df_sample1.reset_index(drop=True)
df.head()
```

```python
df_final = df.replace({'Label' : { 'ham' : 0, 'spam' : 1}})
df_final.head()
```

```python
df_final.to_csv('final_data.csv', index=False)
```
